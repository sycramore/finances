```python
#Calculate capital increase with monthly rates and an initial capital
#Uses a simple exponential function for the inflation and interest rate increase
#coupled with a linear increase of type capital = capital_per_year*years + initial_capital
#This model could be used for an ETF-based savings plan
#The example shown in the accompagning figure shows such a development for an ETF-based savings plan
#with initial capital of 1000 Euro, a monthly rate of 100 Euro, an ETF-typical interest rate of 6% and
#a typical inflation rate of 2 percent over a time of 10 years
#Inflation rates are accounted for in the model by taking as netto interest rate the difference between
#the nominal interest rate and the inflation. This allows for an inflation-adjusted model

import matplotlib.pyplot as plt

print('Lifetime of your portfolio in years (must be an integer number):')
lifetime = int(input())

print('Initial capital:')
ini_cap = float(input())

print('Monthly rate:')

cipm = float(input())

cipa = 12 * cipm

print('Interest rate in percent:')

interest_rate = float(input())

print('Inflation in percent')

inflation = float(input())

def total_cap(ini_cap,lifetime,cipa,evpa):
        return((ini_cap + lifetime*cipa)*(1 + 0.01*evpa)**lifetime)

#total_capital_brutto = [total_cap(ini_cap,lt,cipa,interest_rate) for lt in range(0,lifetime)] #Modell Boerse
#total_capital_after inflation = [gesamtkapital(ini_cap,lt,cipa,inflation) for lt in range(0,lifetime)] #keine verzinsung und 2 Prozent Inflation, Modell Sparbuch
years = [lt for lt in range(0,lifetime)]

total_capital = [total_cap(ini_cap,lt,cipa,(interest_rate-inflation)) for lt in range(0,lifetime)]

plt.plot(years,total_capital)
plt.xlabel('Time in years')
plt.ylabel('Capital in your chosen currency')
plt.savefig('development_of_total_capital.png')
plt.show()

for i in range(lifetime):
    print('The total capital at year ' + str(i) + ' is ' + str(round(total_capital[i])) + ' Euro')
```
